```python
import numpy as np
import pandas as pd
```

```python
book = 'Books.csv'
rating = 'Ratings.csv'
User = 'Users.csv'

books = pd.read_csv(book)
ratings = pd.read_csv(rating)
users = pd.read_csv(User)
```

```python
books.head()
```

```python
ratings_with_name = ratings.merge(books, on = 'ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()[['Book-Rating']].reset_index()
num_rating_df.rename(columns = {'Book-Rating' : 'num_rating'}, inplace= True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
```

```python
popular_df = popular_df[popular_df['num_rating']>=250].sort_values('avg_rating', ascending = False).head(50)
```

```python
popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'avg_rating', 'num_rating']].reset_index()
```

```python
popular_df = popular_df.rename(columns = {'avg_rating' : 'Avg_rating'})
popular_df = popular_df.rename(columns = {'num_rating' : 'Num_rating'})
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>=200
padhe_like_users = x[x].index
padhe_like_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_like_users)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python
 final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):

    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1], reverse = True)[1:25]
    
    
    for i in similar_items:
       print(pt.index[i[0]])
        
```

```python
recommend('Animal Farm')
```
